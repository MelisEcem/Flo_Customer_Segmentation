```python
df=pd.read_csv("/kaggle/input/flo-data/flo_data_20k.csv")
df.head()
```

```python
def data_review (dataframe):
    print("**Shape**")
    print(dataframe.shape)
    print("**Types**")
    print(dataframe.dtypes)
    print("**Describe**")
    print(dataframe.describe().T)
    print("**NA**")
    print(dataframe.isnull().sum())
data_review(df)
```

```python
df["Total_Order"] = df["order_num_total_ever_online"] + df["order_num_total_ever_offline"]
df["Total_Price"] = df["customer_value_total_ever_offline"] + df["customer_value_total_ever_online"]
df.head()
```

```python
date_col = [col for col in df.columns if "date" in col]
print(date_col)
```

```python
for col in date_col:
   df[col]=pd.to_datetime(df[col])
df.dtypes
```

```python
df.groupby("order_channel").agg({"master_id":"count",
                                 "Total_Order":"sum",
                                 "Total_Price":"sum"})
```

```python
df["last_order_date"].max()
```

```python
rfm=pd.DataFrame()
rfm["Customer_Id"] = df["master_id"]
analysis_date = df["last_order_date"].max() + pd.Timedelta(2, "D")
rfm["recency"] = (analysis_date-df["last_order_date"]).dt.days
rfm["frequence"] = df["Total_Order"]
rfm["monetary"] =df["Total_Price"]
rfm.head()
```

```python
rfm["recency_score"]=pd.qcut(rfm["recency"],5,[5,4,3,2,1])
rfm["frequence_score"]=pd.qcut(rfm["frequence"].rank(method="first"),5,[1,2,3,4,5])
rfm["monetary_score"]=pd.qcut(rfm["monetary"],5,[1,2,3,4,5])
rfm.head()
```

```python
rfm["RF_SCORE"]= rfm["recency_score"].astype(str) + rfm["frequence_score"].astype(str)
rfm.head()
```

```python
seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'
}
rfm["segment"]=rfm["RF_SCORE"].replace(seg_map,regex=True)
rfm.head()
```

```python
rfm.groupby("segment").agg({"recency":"mean",
                            "frequence":"mean",
                            "monetary":"mean",})
```

```python
df["segment"]=rfm["segment"]
df.head()
```

```python
woman_customer=df[df["interested_in_categories_12"].str.contains("KADIN")][["master_id","segment"]]
best_woman_customer = woman_customer[(woman_customer["segment"]=="champions") | (woman_customer["segment"]=="loyal_customers")]["master_id"]
print(best_woman_customer)
best_woman_customer.to_csv("new_brand_target_customer.csv",index=False)
```

```python
man_children_customer=df[(df["interested_in_categories_12"].str.contains("ERKEK")) | (df["interested_in_categories_12"].str.contains("OCUK"))][["master_id","segment"]]
customer_for_discount=man_children_customer[(man_children_customer["segment"]=="about_to_sleep") |
                      (man_children_customer["segment"]=="cant_loose") |
                      (man_children_customer["segment"]=="new_customer")]["master_id"]
print(customer_for_discount)
customer_for_discount.to_csv("target_customers_for_discount.csv",index=False)
```
